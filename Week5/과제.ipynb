- ### pandas 라이브러리로 URL 을 활용하여 데이터 가져오기

```python
import pandas as pd
# url = "https://ko.wikipedia.org/wiki/올림픽_메달_집계"
url = "https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84"
# url = https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84
df = pd.read_html(url)
df
```

- ### df[0] 을 활용하여 index 0 에 해당하는 표 가져오기
    -  해당 페이지에 여러 개의 표가 존재 (10 개)

```python
len(df)
```

```python
df[0]
```

- ###  헤더를 열 이름(columns)으로, 나라 이름을 행 이름(index)로 지정

```python
df = pd.read_html(url, header=0, index_col=0)
df[0]
```

- ### 슬라이싱을 활용하여 동계올림픽에 대한 데이터 추출

```python
# 슬라이싱을 활용하여 동계올림픽에 대한 데이터 추출
a = df[0].iloc[:,5:10]
a
```

- ### Unnamed 부분 컬럼명 설정하기

```python
# 컬럼명 설정하기
a.columns = ['동계 참가 횟수', '금', '은', '동', '계']
a
```

- ### 동계 참가 횟수와 금메달 기준으로 내림차순 정렬하기

```python
# 동계 참가 횟수와 금메달 기준으로 내림차순 정렬하기
a = a.sort_values(by=['동계 참가 횟수','금'], ascending=False)
a
```

- ### 해당 결과를 csv 파일로 저장하기

```python
# csv 파일로 저장하기
a.to_csv('winter_medal.csv')
```

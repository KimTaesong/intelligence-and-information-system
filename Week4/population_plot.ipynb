### 우리 동네 연령별 인구수 데이터 시각화 하기
#### zip 함수
- 동일한 개수로 이루어진 자료형을 묶어 주는 역할을 하는 함수
- row[3:104]: 남성 만 0 세 만 100 세 이상 데이터가 담긴 리스트
- 길이 : 101
- row[106:207]: 여성 만 0 세 만 100 세 이상 데이터가 담긴 리스트
- 길이 : 101

```python
import csv

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

result = []
for row in data:
    if '송도1동' in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            result.append(int(m) + int(f))
print(result)
```

```python
import csv

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

result = []                                     # 빈 리스트 만들기
for row in data:
    if '송도1동' in row[0]:                     # 송도1동이 포함된 행적구역 찾기
        for m,f in zip(row[3:104],row[106:207]):# 0세부터 끝(100세 이상)까지 모든 연령에 대해 반복하기
            result.append(int(m) + int(f))      # 해당 연령의 인구수 리스트에 순서대로 저장하기
print(result)                                   # 0세부터 100세 이상까지의 인구수 출력하기
```

```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

result = []
name = input('인구 구조를 알고 싶은 동 이름 입력 : ')
for row in data:
    if name in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            result.append(int(m) + int(f))
# print(result)
plt.rc('font', family = 'Malgun Gothic')
plt.title('송도1동 연령별 인구수')

plt.style.use('ggplot') # 격자 무늬 스타일 지정
plt.plot(result)
plt.xlabel('연령')
plt.ylabel('인구수')
plt.show()
```

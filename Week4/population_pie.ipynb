## 파이 차트로 비율 표현하기
- ### pie() 함수
    - 파이 차트 (pie 는 원그래프 라고도 불림
    - 전체 데이터 중 특정 데이터의 비율을 보기 쉽게 표현
        - 입력한 데이터의 크기에 따라 원의 비율이 결정됨
- plt.pie([10, 20])

```python
import matplotlib.pyplot as plt
plt.pie([10,20])
plt.show()
```

### 동그란 원으로 표현하기
- axis() 함수를 추가하여 동그란 원 그리기 가능
    - plt.axis (‘ 이 없는 경우 , 찌그러진 원이 그려질 수 있음

```python
import matplotlib.pyplot as plt

plt.axis('equal')
plt.pie([100,200,300,400])
plt.show()
```

### 레이블 추가하기
- plt.pie(labels=['레이블 1’,‘레이블 2',‘레이블 3', '레이블 4'])

```python
import matplotlib.pyplot as plt

plt.rc('font', family ='Malgun Gothic')
plt.axis('equal')
plt.pie([100, 200, 300, 400], labels=['레이블1', '레이블2', '레이블3', '레이블4'])
plt.show()
```

### 비율 및 범례 표시하기
- plt.pie(autopct =‘%.1f%%)
- plt.legend()
- 정확한 계산을 위해 소수점까지 나타내는 실수 (float) 형태로 표시
- 소수점 아래 n 번째 자리까지 나타내려면
- 1f 소수점 아래 둘째자리에서 반올림한 값

```python
import matplotlib.pyplot as plt

plt.rc('font', family ='Malgun Gothic')
plt.axis('equal')
size = [120, 250, 330, 410]
label=['레이블1', '레이블2', '레이블3', '레이블4']
plt.pie(size, labels=label,autopct='%.1f%%')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family ='Malgun Gothic')
plt.axis('equal')
size = [120, 250, 330, 410]
label=['레이블1', '레이블2', '레이블3', '레이블4']
plt.pie(size, labels=label,autopct='%.2f%%')
plt.legend()
plt.show()
```

### 색 정하기
- plt.pie(colors=['darkmagenta','deeppink','hotpink','pink'])

```python
import matplotlib.pyplot as plt

plt.rc('font', family ='Malgun Gothic')
plt.axis('equal')
size = [120, 250, 330, 410]
label=['레이블1', '레이블2', '레이블3', '레이블4']
color=['darkmagenta','deeppink','hotpink','pink']
plt.pie(size, labels=label,autopct='%.2f%%',colors=color)
plt.legend()
plt.show()
```

### 돌출 효과로 강조하기
- plt.pie (explode=(0, 0, 0.1,0))
    - 강조하고 싶은 데이터의 값 변경

```python
import matplotlib.pyplot as plt

plt.rc('font', family ='Malgun Gothic')
plt.axis('equal')
size = [120, 250, 330, 410]
label=['레이블1', '레이블2', '레이블3', '레이블4']
color=['darkmagenta','deeppink','hotpink','pink']
exp = (0, 0, 0.1, 0) # 0,1 돌출되는 정도 입력 (0: 돌출되지 않음)
plt.pie(size, labels=label,autopct='%.2f%%',colors=color, explode=exp)
plt.legend()
plt.show()
```

### 돌출 효과로 강조하기
- #### 만약 0.1 -> 0.3 으로 변경한다면

```python
import matplotlib.pyplot as plt

plt.rc('font', family ='Malgun Gothic')
plt.axis('equal')
size = [120, 250, 330, 410]
label=['레이블1', '레이블2', '레이블3', '레이블4']
color=['darkmagenta','deeppink','hotpink','pink']
exp = (0, 0, 0.3, 0) # 0,3 돌출되는 정도 입력 (0: 돌출되지 않음)
plt.pie(size, labels=label,autopct='%.2f%%',colors=color, explode=exp)
plt.legend()
plt.show()
```

## 2. 성별 인구 비율 표현하기
### 성별 연령별 인구 데이터 출력하기
- 성별 인구 총합을 더하여 파이 차트로 그리기

```python
import csv

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

result = []
m_sum, f_sum = 0, 0

for row in data:
    if '송도1동' in row[0]:
        for i in range(101):
            m_sum = m_sum + int(row[i+3])    #   3~103번까지 데이터 더하기
            f_sum += int(row[i+106])         # 106~206번까지 데이터 더하기
            
result.append(m_sum)            # 남성 합계 데이터 리스트에 추가
result.append(f_sum)            # 여성 합계 데이터 리스트에 추가
print(result)
```

### 피이 차트로 표현하기
- plt.pie(startangle =90) 이용하기
- startangle 속성은 파이 차트의 시작 각도를 정해줌
- 90 으로 설정한 경우 반시계방향으로 90 도 이동하여, 12 시 정각 위치에서 그래프가 시작됨

```python
import matplotlib.pyplot as plt

plt.rc('font', family ='Malgun Gothic')
plt.axis('equal')

label = ['남', '여']
color = ['crimson', 'darkcyan']
plt.pie(result, labels=label, autopct='%.1f%%',colors=color, startangle=90)
plt.legend()
plt.show()
```

## 1. 꺽은선 그래프로 표현하기
- 성별 연령별 인구 데이터 저장하기

```python
import csv

f = open('incheon_population.csv','r')
data = csv.reader(f)

male,female = [], []
name = input('궁금한 동네 이름을 입력해주세요 : ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            male.append(int(row[i]))         # 남성 데이터 저장하기
            female.append(int(row[i+103]))   # 여성 데이터 저장하기
        break
```

## 성별 연령별 인구 데이터 꺾은선 그래프로 표현하기
- 어떻게 하면 연령대별 성별 비율을 더 알아보기 쉽게 나타낼 수 있을까?

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.title(name + '연령별 성별 인구수')
plt.plot(male, label='남')
plt.plot(female, label='여')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.legend()
plt.show()
```

## 막대그래프로 표현하기
- ### 연령별 (남성 인구-여성 인구) 값 result 에 저장하기

```python
import csv
f = open('incheon_population.csv','r')
data = csv.reader(f)

# male, female = [], []
result = []
name = input('궁금한 동네 이름을 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
           m = int(row[i])         # 남성 데이터 저장하기
           f = int(row[i+103])   # 여성 데이터 저장하기
           result.append(m-f)
        break
```

### 연령별 (남성 인구 -  여성 인구) 값 막대그래프로 표현하기

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 연령별 성별 인구수')
plt.bar(range(101),result)
plt.xlabel('나이')
plt.ylabel('남성 인구 - 여성 인구')
plt.show()
```

### 같은 데이터라도 다양한 형태로 시각화하면 데이터에 숨겨진 다양한 사실을 발견할 수 있음
- 항아리 모양의 그래프
- 꺾은선 그래프
- 막대그래프

## 3. scatter( ) 함수로 표현하기
### 산점도 점이 흩어져있는 그래프 로 표현하기
- 'scatter’ 흩뿌리다 , 사방으로 흩어지다 ’-> 여기저기 점이 흩어진 모양의 그래프
- x 축과 y 축에 해당하는 <b>데이터의 상관관계</b> 를 표현할 때 사용
    - 두 개의 축을 기준으로 데이터가 얼마나 퍼져있는지(분포) 알 수 있어 산포도 라고도 불림

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.scatter([1,2,3,4],[10,30,20,40])
plt.show()
```

## 4. 버블 차트로 표현하기
- ### 버블 차트로 표현하기
    - scatter() 함수를 이용하여 버블 차트를 그릴 수 있음
        - 기존 코드에서 size 를 의미하는 s 속성을 추가하고 , 원하는 크기를 입력

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.scatter([1,2,3,4],[10,30,20,40],s=[100,200,250,300])
plt.show()
```

### 버블 차트로 표현하기 컬러 추가

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
x_list = [1,2,3,4]
y_list = [10,30,20,40]
size = [100,200,250,300]
plt.scatter(x_list, y_list, s=size, c=['r','b','g','gold'])
plt.show()
```

### 버블 차트로 표현하기 컬러바 추가
- colorbar () 함수 이용
- scatter() 함수에 <b>c 속성</b>을 추가하여 , 표현하고 싶은 <b>색상의 개수 설정</b><br> --> 각 데이터에 해당하는 컬러바의 색으로 결정됨

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
x_list = [1,2,3,4]
y_list = [10,30,20,40]
size = [100,200,250,300]
plt.scatter(x_list, y_list, s=size, c=range(4))
plt.colorbar()
plt.show()
```

### 버블 차트로 표현하기 컬러맵 추가
- 무지개색과 비슷한 jet 컬러맵


```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
x_list = [1,2,3,4]
y_list = [10,30,20,40]
size = [100,200,250,300]
plt.scatter(x_list, y_list, s=size, c=range(4),cmap='jet')
plt.colorbar()
plt.show()
```

## 막대 그래프
- 꺾은선 그래프가 아닌 다른 형태로 시각화하면 어떤새로운 정보들을 찾을 수 있을까
- bar() 함수
- bar(막대를 표시할 위치 , 막대의 높이 길이)

```python
import matplotlib.pyplot as plt

plt.bar([0,1,2,4,6,10],[1,2,3,5,6,7])
plt.show()
```

## bar() 함수
- bar(막대를 표시할 위치 , 막대의 높이)
- range() 함수를 이용하여 오름차순으로 그리기
- range(6)  range(0, 6) 과 동일 →[0, 1, 2, 3, 4, 5]

```python
import matplotlib.pyplot as plt

plt.bar(range(6),[1,2,3,5,6,7])
plt.show()
```

### 막대그래프 를 활용해 우리 동네 인구 구조 시각화하기
- 송도 1 동 이 아닌 다른 동이 궁금하다면 , 해당 부분을 수정

```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

result = []
for row in data:
    if '송도1동' in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            result.append(int(m) + int(f))
# print(result)
plt.rc('font', family = 'Malgun Gothic')
plt.title('송도1동 연령별 인구수')

plt.bar(range(101),result)
plt.xlabel('연령')
plt.ylabel('인구수')
plt.show()
```

### 수평 막대그래프 barh () 함수 를 활용해 우리 동네 인구구조 시각화하기
- plt.barh (range(101),result)
    - range(101): y 축의 막대 위치
    - result: 막대의 너비

```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

result = []
for row in data:
    if '송도1동' in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            result.append(int(m) + int(f))
# print(result)
plt.rc('font', family = 'Malgun Gothic')
plt.title('송도1동 연령별 인구수')

plt.barh(range(101),result)
plt.xlabel('연령')
plt.ylabel('인구수')
plt.show()
```

## 항아리 모양 그래프 그리기
- ### 성별 데이터 저장하기

```python
import csv

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

male = []
female = []
for row in data:
    if '송도1동' in row[0]:
        for i in row[3:104]:
            male.append(int(i))   # 남성 데이터를 리스트 male에 저장
        for i in row[106:207]:
            female.append(int(i)) # 여성 데이터를 리스트 female에 저장
```

- ### 성별 데이터 수평 막대그래프로 시각화하기

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.title('송도1동 연령별 인구수')

plt.barh(range(101),male)
plt.barh(range(101),female)
plt.xlabel('연령')
plt.ylabel('인구수')
plt.show()
```

- ### 성별 데이터 수평 막대그래프로 시각화하기
    - 남성 데이터 음수로 바꾸기

```python
import csv

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

male = []
female = []
for row in data:
    if '송도1동' in row[0]:
        for i in row[3:104]:
            # male.append(int(i))   # 남성 데이터를 리스트 male에 저장
            male.append(-int(i))   # 마이너스 부호를 넣어서 음수로 변경
        for i in row[106:207]:
            female.append(int(i)) # 여성 데이터를 리스트 female에 저장
```

- ### 성별 데이터 수평 막대그래프로 시각화하기
    - 마이너스 부호 깨짐 해결
    - 범례 넣기

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.title('송도1동 연령별 인구수')

plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101),male, label ='남성')
plt.barh(range(101),female, label = '여성')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.legend()
plt.show()
```
